```python
import csv
import time
import io
import json
import os
from datetime import datetime
import matplotlib.pyplot as plt

# -----------------------------
# Configurable Paths
# -----------------------------
PATIENTS_DB = 'patients.json'
OUTPUT_CSV = 'therapy_output.csv'

# -----------------------------
# EEG Data Source (CSV or Simulated)
# -----------------------------
# Unique EEG Data per Patient
# -----------------------------
# EEG Data Source (CSV or Simulated)
# -----------------------------
# Unique EEG Data per Patient
patient_eeg_data = {
    "P001": """alpha,beta,theta
0.45,0.72,0.18
0.66,0.52,0.15
0.39,0.48,0.44
0.53,0.57,0.25
0.62,0.59,0.12""",
    "P002": """alpha,beta,theta
0.55,0.68,0.20
0.61,0.57,0.18
0.42,0.44,0.40
0.58,0.60,0.22
0.63,0.62,0.10""",
    "P003": """alpha,beta,theta
0.35,0.80,0.10
0.30,0.75,0.20
0.25,0.72,0.30
0.33,0.77,0.25
0.40,0.74,0.15""",  # Removed comment
    "P004": """alpha,beta,theta
0.30,0.75,0.22
0.28,0.80,0.20
0.22,0.78,0.25
0.35,0.77,0.30
0.40,0.76,0.18""",  # Removed comment
    "P005": """alpha,beta,theta
0.52,0.70,0.18
0.63,0.53,0.17
0.43,0.46,0.43
0.54,0.58,0.23
0.60,0.61,0.11""",
    "P006": """alpha,beta,theta
0.62,0.74,0.16
0.65,0.56,0.14
0.44,0.49,0.41
0.57,0.59,0.20
0.63,0.64,0.12""",
    "P007": """alpha,beta,theta
0.50,0.67,0.19
0.62,0.53,0.16
0.40,0.45,0.42
0.55,0.58,0.25
0.61,0.62,0.13""",
    "P008": """alpha,beta,theta
0.49,0.73,0.20
0.60,0.54,0.18
0.42,0.47,0.39
0.53,0.56,0.24
0.59,0.60,0.15""",
    "P009": """alpha,beta,theta
0.53,0.71,0.17
0.64,0.55,0.16
0.45,0.48,0.40
0.56,0.59,0.22
0.62,0.63,0.12""",
    "P010": """alpha,beta,theta
0.51,0.68,0.19
0.60,0.52,0.17
0.43,0.47,0.41
0.54,0.57,0.23
0.59,0.60,0.14"""
}

state_to_int = {"stressed": 3, "focused": 2, "fatigued": 1, "neutral": 0}
int_to_state = {v: k for k, v in state_to_int.items()}

# -----------------------------
# Load EEG data
# -----------------------------
def load_eeg_data(patient_id):
    eeg_csv = patient_eeg_data.get(patient_id, "").strip()
    if eeg_csv:
        reader = csv.DictReader(io.StringIO(eeg_csv))
        return [
            {"alpha": float(r['alpha']), "beta": float(r['beta']), "theta": float(r['theta'])}
            for r in reader
        ]
    return []

# -----------------------------
# Cognitive Analysis
# -----------------------------
def evaluate_cognitive_state(eeg):
    if eeg['beta'] > 0.7 and eeg['alpha'] < 0.4:
        return "stressed"
    elif eeg['alpha'] > 0.6 and eeg['theta'] < 0.2:
        return "focused"
    elif eeg['theta'] > 0.4:
        return "fatigued"
    return "neutral"

# -----------------------------
# Neurofeedback Response
# -----------------------------
def neurofeedback_response(state):
    return {
        "stressed": "Apply calming light and breathing audio",
        "fatigued": "Trigger mild vibration and alert sounds",
        "focused": "Reinforce state with soft ambient sounds",
        "neutral": "Maintain baseline stimulation"
    }[state]

# -----------------------------
# Save Patient Profile
# -----------------------------
def load_patients():
    if os.path.exists(PATIENTS_DB):
        with open(PATIENTS_DB, 'r') as f:
            return json.load(f)
    return {}

def save_patients(data):
    with open(PATIENTS_DB, 'w') as f:
        json.dump(data, f, indent=2)

# -----------------------------
# View Patient History
# -----------------------------
def view_patient_history(patient_id):
    patients = load_patients()
    if patient_id not in patients:
        print(f"No record found for patient ID {patient_id}")
        return

    patient = patients[patient_id]
    print(f"Patient: {patient['name']} (ID: {patient_id})")
    for session in patient['history']:
        print(f"\nðŸ“… Date: {session['date']}")
        for result in session['results']:
            print(f"  Cycle {result['cycle']}: {result['state']} ({result['response']})")

# -----------------------------
# Main Therapy Session Logic
# -----------------------------
def run_therapy_session(patient_id, name):
    eeg_data_list = load_eeg_data(patient_id)
    results = []
    state_counts = {"stressed": 0, "focused": 0, "fatigued": 0, "neutral": 0}
    print(f"\nðŸ§  Starting Session for {name} (ID: {patient_id})\n")

    for i, eeg in enumerate(eeg_data_list):
        state = evaluate_cognitive_state(eeg)
        feedback = neurofeedback_response(state)
        print(f"Cycle {i+1} | EEG: {eeg} | State: {state} | Response: {feedback}")
        results.append({"cycle": i+1, **eeg, "state": state, "response": feedback})
        state_counts[state] += 1
        time.sleep(1)

    with open(OUTPUT_CSV.replace('.csv', f'_{patient_id}.csv'), 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=results[0].keys())
        writer.writeheader()
        writer.writerows(results)

    # Save to patient profile
    patients = load_patients()
    if patient_id not in patients:
        patients[patient_id] = {"name": name, "history": []}
    patients[patient_id]["history"].append({
        "date": datetime.now().strftime('%Y-%m-%d'),
        "results": results
    })
    save_patients(patients)

    # Create Subplots for Pie Chart and Line Graph
    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    # Line Graph
    states = [state_to_int[r['state']] for r in results]
    axs[0].plot(range(1, len(states)+1), states, marker='o')
    axs[0].set_title(f"Cognitive State - {name}")
    axs[0].set_xlabel("Cycle")
    axs[0].set_ylabel("State")
    axs[0].set_yticks(list(int_to_state.keys()))
    axs[0].set_yticklabels(list(int_to_state.values()))
    axs[0].grid()
    axs[0].set_xticks(range(1, len(states)+1))

    # Pie Chart
    axs[1].pie(state_counts.values(), labels=state_counts.keys(), autopct='%1.1f%%', startangle=90)
    axs[1].set_title(f"Cognitive State Distribution - {name}")
    axs[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Show the plots
    plt.tight_layout()
    plt.show()

    print(f"\nâœ… Session complete. Results saved to {OUTPUT_CSV.replace('.csv', f'_{patient_id}.csv')} and profile updated.")

# -----------------------------
# Run the System for Multiple Patients
# -----------------------------
if __name__ == '__main__':
    patients_list = [
        ("P001", "Aarav Sharma"),
        ("P002", "Kavya Iyer"),
        ("P003", "Rohan Verma"),
        ("P004", "Neha Kapoor"),
        ("P005", "Vivaan Deshmukh"),
        ("P006", "Ananya Reddy"),
        ("P007", "Devansh Joshi"),
        ("P008", "Priya Nair"),
        ("P009", "Ishaan Gupta"),
        ("P010", "Tanya Singh")
    ]

    for pid, name in patients_list:
        run_therapy_session(pid, name)

    # Optional: View histories for all patients
    for pid, _ in patients_list:
        view_patient_history(pid)
```

```python
import json
import os

# Path to the patient database
PATIENTS_DB = 'patients.json'

# Function to load patients from the JSON file
def load_patients():
    if os.path.exists(PATIENTS_DB):
        with open(PATIENTS_DB, 'r') as f:
            return json.load(f)
    return {}

# Function to view a specific patient's history
def view_patient_history(patient_id):
    patients = load_patients()
    if patient_id not in patients:
        print(f"No record found for patient ID {patient_id}")
        return

    patient = patients[patient_id]
    print(f"Patient: {patient['name']} (ID: {patient_id})")
    for session in patient['history']:
        print(f"\nðŸ“… Date: {session['date']}")
        for result in session['results']:
            print(f"  Cycle {result['cycle']}: {result['state']} ({result['response']})")

# Function to view all patients' data
def view_all_patients():
    patients = load_patients()
    if not patients:
        print("No patients found.")
        return

    for patient_id, patient in patients.items():
        print(f"Patient ID: {patient_id}")
        print(f"Name: {patient['name']}")
        print("History:")
        for session in patient['history']:
            print(f"  ðŸ“… Date: {session['date']}")
            for result in session['results']:
                print(f"    Cycle {result['cycle']}: {result['state']} ({result['response']})")
            print("-" * 50)

# Main function to handle user input and call appropriate functions
def main():
    while True:
        print("\nWelcome to the Patient Data System")
        print("1. View specific patient's data")
        print("2. View all patients' data")
        print("3. Exit")

        choice = input("Please choose an option (1, 2, 3): ")

        if choice == '1':
            # Get patient ID from user
            patient_id = input("Enter the Patient ID (e.g., P001): ").strip()
            view_patient_history(patient_id)

        elif choice == '2':
            # View all patients
            view_all_patients()

        elif choice == '3':
            # Exit the program
            print("Exiting the system. Goodbye!")
            break

        else:
            print("Invalid option. Please try again.")

# Run the program
if __name__ == '__main__':
    main()
```
